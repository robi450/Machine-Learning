```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler


# Set numpy print options to enhance readability
np.set_printoptions(precision=2)

# Print to confirm everything is set up correctly
print("Libraries imported successfully!")
```

```python
# Create target data (quadratic function)
x = np.arange(0, 20, 1) # values from 0 to 19
y = 1 + x**2 # Define the quadratic relationsip

# Reshape x to a 2D array for consistency with sklearn models
X = x.reshape(-1, 1)

# Plot the original data to visualize
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Initial Target Data (Quadratic Relationship)")
plt.legend()
plt.show()

```

```python
# Create mock dataset for housing prices
# Area in square feet, and prices in thousands of dollars
area = np.array([500, 750, 1000, 1200, 1500, 1800, 2000, 2500, 3000, 3500])
prices = np.array([150, 200, 240, 310, 400, 450, 500, 600, 750, 850])

# Reshape area to a 2D array for sklearn models
X_area = area.reshape(-1, 1)

# Plotting the data
plt.scatter(area, prices, color='blue', marker='o', label="Actual Housing Prices")
plt.xlabel("Area (square feet)")
plt.ylabel("Price (in thousands of dollars)")
plt.title("Housing Prices vs Area")
plt.legend()
plt.show()
```

```python
# Fit a Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_area, prices)

# Predict housing prices using linear regression
prices_pred_linear = linear_model.predict(X_area)

# Plot the linear regression line against the data
plt.scatter(area, prices, color='blue', marker='o', label="Actual Housing Prices")
plt.plot(area, prices_pred_linear, color='red', label="Linear Regression Fit")
plt.xlabel("Area (square feet)")
plt.ylabel("Price (in thousands of dollars")
plt.title("Linear Regression Fit to Housing Prices")
plt.legend()
plt.show()


```

```python
# Engineer quadratic features (degree 2)
poly_features = PolynomialFeatures(degree=2)
X_area_poly = poly_features.fit_transform(X_area) # add x^2 as a feature

# Fit a polynomial Regression model (degree 2)
poly_model = LinearRegression()
poly_model.fit(X_area_poly, prices)

# Predict housing prices using the polynomial model
prices_pred_poly = poly_model.predict(X_area_poly)

# Plot the quadratic fit against the data
plt.scatter(area, prices, color='blue', marker='o', label="Actual Housing Prices")
plt.plot(area, prices_pred_poly, color='green', label="Polynomial Regression (Degree 2)")
plt.xlabel("Area (square feet)")
plt.ylabel("Price (in thousands of dollars")
plt.title("Polynomial Regression Fit to Housing Prices")
plt.legend()
plt.show()


```

```python
# Engineer cubic features (degree 3)
poly_features_high = PolynomialFeatures(degree=3)
X_area_poly_high = poly_features_high.fit_transform(X_area)  # Add x^2 and x^3 as features

# Fit a Polynomial Regression model (degree 3)
poly_model_high = LinearRegression()
poly_model_high.fit(X_area_poly_high, prices)

# Predict housing prices using the polynomial model with cubic features
prices_pred_poly_high = poly_model_high.predict(X_area_poly_high)

# Plot the cubic fit against the data
plt.scatter(area, prices, color='blue', marker='o', label="Actual Housing Prices")
plt.plot(area, prices_pred_poly_high, color='orange', label="Polynomial Regression (Degree 3)")
plt.xlabel("Area (square feet)")
plt.ylabel("Price (in thousands of dollars)")
plt.title("Cubic Polynomial Regression Fit to Housing Prices")
plt.legend()
plt.show()
```

```python
# Scale features using StandardScaler (Z-score normalization)
scaler = StandardScaler()
X_area_poly_high_scaled = scaler.fit_transform(X_area_poly_high)

# Fit a Linear Regression model on the scaled features
scaled_model = LinearRegression()
scaled_model.fit(X_area_poly_high_scaled, prices)

# Predict housing prices using the scaled model
prices_pred_scaled = scaled_model.predict(X_area_poly_high_scaled)

# Plot the results with scaled features
plt.scatter(area, prices, color='blue', marker='o', label="Actual Housing Prices")
plt.plot(area, prices_pred_scaled, color='purple', label="Polynomial Regression with Scaled Features")
plt.xlabel("Area (square feet)")
plt.ylabel("Price (in thousands of dollars)")
plt.title("Polynomial Regression with Scaled Features")
plt.legend()
plt.show()
```
